### 3. Membaca file dari gambar dan menampilkannya





```python
from google.colab.patches import cv2_imshow 
image = cv2.imread('ranger-green-in-air1.png') 

cv2_imshow(image)
```

### 4. Teknik mengolah gambar dengan mengonversi warna BGR ke RGB










```python
# selanjutnya mengolah gambar 
import matplotlib.pyplot as plt
fig = plt.figure() 
fig.add_subplot(121) # ini memasukkan jumlah baris dan kolom 

plt.imshow(image) 

fig.add_subplot(122) 

image2 = cv2.cvtColor(image, cv2.COLOR_BGR2RGB) # menggunakan colorRGB yang ditukar warnanya sehingga output warna power ranger tosca-orange berubah menjadi green-blue 

plt.imshow(image2) 

plt.show()
```

### 5. Mengecek dimensi dari gambar yang digunakan

```python
b, g, r = cv2.split(image) 

print('dimensi gambar:')
print(image.shape)
```

6. Teknik mendapatkan dan mengatur piksel gambar yang terdiri dari 3 (tiga) intensitas red, green, blue.

```python
image3 = image 
for i, row in enumerate(image): 
  for j, pixel in enumerate(image): 
    means = (b[i][j]+g[i][j]+r[i][j])/3 
    image3[i][j] = [means, means, means] 

cv2_imshow(image3) 
```

### 7. Teknik melakukan filtering dan menambahkan intensitas gambar 

```python
# teknik filtering 
image4 = cv2.imread('ranger-green-in-air1.png') 
cv2_imshow(image4) 
r, g, b = cv2.split(image4) 

for i, row in enumerate(image4): 
  for j, pixel in enumerate(image4): 
    if r[i][j] <= 50: 
      r1 = 0
      g1 = 0 
      b1 = 0 
    else: 
      r1 = 255 
      g1 = 255 
      b1 = 255 
    image4[i][j] = [r1, g1, b1] 

cv2_imshow(image4) 

for i, row in enumerate(image4): 
  for j, pixel in enumerate(image4): 
    if r[i][j] <= 50 + 70: 
      r1 = 0
      g1 = 0 
      b1 = 0 
    else: 
      r1 = 255 
      g1 = 255 
      b1 = 255 
    image4[i][j] = [r1, g1, b1] 

cv2_imshow(image4)
```

### 8. Teknik membalik proses dan membalik gambar dengan flip

```python
image5 = cv2.imread('ranger-green-in-air1.png')
image_flip = cv2.flip(image5, 1) 
cv2.imwrite('ranger-green-in-air1.png', image_flip) 
cv2_imshow(image5)

image5 = cv2.imread('ranger-green-in-air1.png') 
cv2_imshow(image5)  
```
