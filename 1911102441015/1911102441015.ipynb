```python
from google.colab.patches import cv2_imshow # memanggil fungsi google colab untuk perbaikan cv2.imshow
img = cv2.imread('pak luhut.jpg') # menyimpan gambar dengan fungsi imread dari OpenCV

cv2_imshow(img) # menampilkan gambar dengan fungsi cv2_imshow
```

```python
#convert color
# membuat sebuah variabel kosong 'fig'
fig = plt.figure()
fig.add_subplot(121) # subplot (jumlah baris, jumlah kolom, urutan)

plt.imshow(img) # menampilkan gambar dengan fungsi plt

fig.add_subplot(122) ## subplot, formatnya: (jumlah baris, jumlah kolom, urutan)

img2 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) 

plt.imshow(img2)

plt.show() #untuk menampilkan plot
```

```python
factor1 = PIL.Image.open('pak luhut.jpg')
# buat instansi enhancer untuk brightness 
converter = ImageEnhance.Brightness(img)  

# Plot image 
plt.gcf().set_size_inches(15,15) 
plt.subplot(131),plt.imshow(factor1),plt.title('Original')
# Membuat cahaya lebih gelap
factor2 = converter.enhance(0.5)  
plt.xticks([]), plt.yticks([]) 
plt .subplot(132),plt.imshow(factor2),plt.title('darkens')
# Membuat cahaya lebih terang
factor3 = converter.enhance(2) 
plt.xticks([]), plt.yticks([]) 
plt.subplot(133),plt.imshow(factor3),plt.title('Brightness') 
plt.xticks([]), plt.yticks([]) 

plt.imshow(factor1)
plt.imshow(factor2)
plt.imshow(factor3)
```

```python
img = cv2.imread('pak luhut.jpg',0)
ret,thresh1 = cv2.threshold(img,127,255,cv2.THRESH_BINARY)
ret,thresh2 = cv2.threshold(img,127,255,cv2.THRESH_BINARY_INV)
ret,thresh3 = cv2.threshold(img,127,255,cv2.THRESH_TRUNC)
ret,thresh4 = cv2.threshold(img,127,255,cv2.THRESH_TOZERO)
ret,thresh5 = cv2.threshold(img,127,255,cv2.THRESH_TOZERO_INV)

titles = ['Original Image','BINARY','BINARY_INV','TRUNC','TOZERO','TOZERO_INV']
images = [img, thresh1, thresh2, thresh3, thresh4, thresh5]

for i in range(4):
    plt.subplot(2,3,i+1),plt.imshow(images[i],'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])

plt.show()
```

```python
#flip image
img = cv2.imread('pak luhut.jpg')
print(type(img))

print(img.shape)

img_flip_ud = cv2.flip(img, 0)
cv2.imwrite('pak luhut_flip_ud.jpg', img_flip_ud)

img_flip_lr = cv2.flip(img, 1)
cv2.imwrite('pak luhut_flip_lr.jpg', img_flip_lr)

img0 = cv2.imread('pak luhut.jpg')
img4 = cv2.imread('pak luhut_flip_ud.jpg')
img5 = cv2.imread('pak luhut_flip_lr.jpg')

cv2_imshow(img0)
cv2_imshow(img4)
cv2_imshow(img5)
```
